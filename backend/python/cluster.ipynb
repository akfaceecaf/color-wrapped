```python
import json
from sklearn.cluster import HDBSCAN, KMeans
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from PIL import Image
import requests
from io import BytesIO
import math
```

```python
def to_encoded_hsl(rgba):
    if rgba is None or len(rgba) < 3 or rgba[3] == 0:
        return None
    
    r, g, b = rgba[0] / 255, rgba[1] / 255, rgba[2] / 255
    
    max_c = max(r, g, b)
    min_c = min(r, g, b)
    l = (max_c + min_c) / 2
    c = max_c - min_c
    
    if c == 0:
        return [0, 0, l]
    
    s = c / (max_c + min_c) if l <= 0.5 else c / (2 - max_c - min_c)
    
    if max_c == r:
        h = (g - b) / c + (6 if g < b else 0)
    elif max_c == g:
        h = (b - r) / c + 2
    else:
        h = (r - g) / c + 4
    
    h *= 60
    h_rad = h * np.pi / 180
    
    return [np.sin(h_rad) * s, np.cos(h_rad) * s, l]
```

```python
mock = pd.DataFrame(raw)
mock = mock[['track_name','image_url','average_color']]
mock['hsl'] = mock['average_color'].apply(to_encoded_hsl)
```

```python
vectors = np.array(mock['hsl'].to_list())
colors = np.array(mock['average_color'].to_list())
urls = np.array(mock['image_url'])
names = np.array(mock['track_name'])
```

```python
# configs
configs = {
    # 'km_8': {'algorithm': 'kmeans', 'n_clusters': 8},
    # 'hdb_2_1_leaf': {'algorithm': 'hdbscan', 'min_cluster_size': 2, 'min_samples':1, 'cluster_selection_method': 'leaf'},
    'hdb_2_1_eom': {'algorithm': 'hdbscan', 'min_cluster_size': 2, 'min_samples':1},
}
```

```python
def cluster(vectors, config):
    if config['algorithm'] == 'kmeans':
        model = KMeans(n_clusters=config.get('n_clusters', 8),random_state=42)
    elif config['algorithm'] == 'hdbscan':
        model = HDBSCAN(min_cluster_size=config.get('min_cluster_size', 3), min_samples=config.get('min_samples',2), cluster_selection_method=config.get('cluster_selection_method','eom'))
    return model.fit_predict(vectors)
```

```python
def load_image(url):
    response = requests.get(url)
    return Image.open(BytesIO(response.content))
```

```python
def visualize_cluster(labels, songs:pd.DataFrame, show_image:bool):
    clusters = set(labels)
    for cluster in clusters:
        mask = labels == cluster
        cluster_songs = songs[mask]
        n = len(cluster_songs)
        ncols = 4
        nrows = math.ceil(n / ncols)

        fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(20, 3 * nrows))
        axes = np.array(axes).flatten()

        for i, (_,song) in enumerate(cluster_songs.iterrows()):
            ax = axes[i]
            url = song['image_url']
            color = np.array(song['average_color'],dtype=float) / 255
            if show_image == True:     
                try:
                    img = load_image(url)
                    ax.imshow(img)
                except:
                    ax.set_facecolor(color)
            else:
                ax.set_facecolor(color)
        fig.suptitle(f'Cluster {cluster}')

        plt.tight_layout()
        plt.show()
```

```python
show_image = True
for name, config in configs.items():
    print(name)
    labels = cluster(vectors, config)
    visualize_cluster(labels, mock, show_image)
```
